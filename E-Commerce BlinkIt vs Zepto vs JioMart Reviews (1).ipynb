```python
import pandas as pd
```

```python
dataset["date"] = pd.to_datetime(dataset["date"],errors='coerce')
dataset.dropna(subset=["date"],inplace=True)
```

```python
# EDA
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(dataset["rating"],bins=5 , kde=True)
plt.xlabel("Rating")
plt.ylabel ("Frequency")
plt.title("Distribution of Ratings")
plt.show()
```

```python
sns.countplot(x="platform" , data=dataset)
plt.title("Number of Reviews per Platform")
plt.xlabel("Platform")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
# Sentiment Analysis
```

```python
dataset["rating"].unique()
```

```python
def sentiment_label(rating):
    if rating >=4:
        return "Positive"
    elif rating ==3:
        return "Natural"
    else :
        return "Nagative"
        
dataset["sentiment"] = dataset["rating"].apply(sentiment_label)
```

```python
sns.countplot(x='sentiment', data=dataset)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
sns.countplot(data=dataset,x='platform',hue='rating',palette='viridis')
plt.title('Rating Distribution per Platform')
plt.xlabel('Platform')
plt.ylabel('rating')
```
